```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences

# 1. Load the IMDb dataset
# num_words specifies the number of most frequent words to keep
max_features = 10000  # vocabulary size
maxlen = 500         # sequence length

print(f"Loading data with max_features={max_features} and maxlen={maxlen}...")
(input_train, y_train), (input_test, y_test) = imdb.load_data(num_words=max_features)

print(f"Train sequences: {len(input_train)}")
print(f"Test sequences: {len(input_test)}")

# 2. Preprocess the data (pad sequences)
print("Padding sequences...")
input_train = pad_sequences(input_train, maxlen=maxlen)
input_test = pad_sequences(input_test, maxlen=maxlen)

print(f"input_train shape: {input_train.shape}")
print(f"input_test shape: {input_test.shape}")

# 3. Build the RNN model
print("Building the model...")
embedding_dim = 32

model = Sequential()
model.add(Embedding(max_features, embedding_dim))
model.add(SimpleRNN(32))
model.add(Dense(1, activation='sigmoid'))

model.summary()

# 4. Compile the model
print("Compiling the model...")
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['acc'])

# 5. Train the model
print("Training the model...")
history = model.fit(input_train, y_train,
                    epochs=10,
                    batch_size=128,
                    validation_split=0.2)

# 6. Evaluate the model
print("Evaluating the model...")
loss, accuracy = model.evaluate(input_test, y_test)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
# @title
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense , Dropout
from tensorflow.keras.preprocessing.sequence import pad_sequences

# 1. Load the IMDb dataset
# num_words specifies the number of most frequent words to keep
max_features = 10000  # vocabulary size
maxlen = 500         # sequence length

print(f"Loading data with max_features={max_features} and maxlen={maxlen}...")
(input_train, y_train), (input_test, y_test) = imdb.load_data(num_words=max_features)

print(f"Train sequences: {len(input_train)}")
print(f"Test sequences: {len(input_test)}")

# 2. Preprocess the data (pad sequences)
print("Padding sequences...")
input_train = pad_sequences(input_train, maxlen=maxlen)
input_test = pad_sequences(input_test, maxlen=maxlen)

print(f"input_train shape: {input_train.shape}")
print(f"input_test shape: {input_test.shape}")

# 3. Build the RNN model
print("Building the model...")
embedding_dim = 32

model = Sequential()
model.add(Embedding(max_features, embedding_dim))
model.add(LSTM(32))
model.add(Dropout(0.3))
model.add(Dense(1, activation='sigmoid'))

model.summary()

# 4. Compile the model
print("Compiling the model...")
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['acc'])

# 5. Train the model
print("Training the model...")
history = model.fit(input_train, y_train,
                    epochs=10,
                    batch_size=128,
                    validation_split=0.2)

# 6. Evaluate the model
print("Evaluating the model...")
loss, accuracy = model.evaluate(input_test, y_test)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```
